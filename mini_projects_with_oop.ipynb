```python
#task-(Food cart order counter):
class Foodie:
  def __init__(self,name):
    self.menu=menu
    self.name=name
    self.order_count=0
    self.item=[]
    self.bill =0
  def order(self,*args):
    for item in args:
      item,quantity = item.split('-')
      self.order_count += 1
      self.item.append(item)
      total_price= menu[item] * int(quantity)
      self.bill += total_price
      print(f'Ordered - {item}, quantity - {quantity}, price (per Unit) - {menu[item]}.')
      print(f'Total price-{total_price}')
  def pay_tips(self,tip=0):
    if tip != 0:
      self.bill += tip
      print(f'Gives {tip}/- tips to the waiter.')
    else:
      print(f'No tips to the waiter.')
  def show_orders(self):
    output = f'{self.name} has {self.order_count} items in the cart.\n'
    output += f'Iteams:{self.item}\n'
    output += f'Total spent:{self.bill}'
    return output


menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}

f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())



```

```python
#task-(online Farm manager project through python OOP):
class  Farmer:
  def __init__(self,arg=None):
    if type(arg) == str:
      self.arg = arg
      print(f'Welcome to your farm, {self.arg}!')
    elif type(arg)== int:
      self.id= arg
      print(f'Welcome to your farm. Your farm ID is {self.id}! ')
    else:
      print('Welcome to your farm!')
    self.crops=[]
    self.fishes=[]

  def addCrops(self,*args):
    for item in args:
      self.crops.append(item)
    if len(args) != 0 :
      print(f'{len(args)} crop(s) added.')
    else:
      print(f'No crop(s) added.')

  def addFishes(self,*args):
    for item in args:
      self.fishes.append(item)
    if len(args) != 0:
      print(f'{len(args)} fish(s) added.')
    else:
      print('No fishe(s) added.')

  def showGoods(self):
    if len(self.crops) == 0 and len(self.fishes) >0 :
      print(f"You don't have any crop(s).")
      print(f'You have {len(self.fishes)} fish(s):')
      print(f'{",".join(self.fishes)}')
    if len(self.crops) >0 and len(self.fishes) == 0 :
      print(f'You have {len(self.crops)} crop(s):')
      print(f'{",".join(self.crops)}')
      print("You don't have any fish(s).")
    if len(self.crops) >0 and len(self.fishes) > 0 :
      print(f'You have {len(self.crops)} crop(s): ')
      print(f'{",".join(self.crops)}')
      print(f'You have {len(self.fishes)} fish(s):')
      print(f'{",".join(self.fishes)}')
f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#task-(online Amusement park ride handler):
class UniversalStudiosUser:
  def __init__(self,name,age,catagory):
    self.name=name
    self.age=age
    self.catagory=catagory
    self.rides={}
    self.bill= 0.0
    print('Welcome to Universal Studios.')

  def selected_rides(self,*args):
    for item in args:
      ride,bill = item.split('-')
      self.rides[ride] = int(bill)
      self.bill += int(bill)
    print('Added ride(s) successfully.')

  def show_details(self):
    print('Your information:')
    print(f'Name:{self.name},Age:{self.age},Catagory:{self.catagory}')
    for ride,bill in self.rides.items():
      print(f'Ride:{ride},Amount:{self.rides[ride]} doller(s)')
    if self.catagory == "Special" and len(self.rides.items()) > 3 :
      self.bill -= self.bill*0.2
      print("Congratulations!!! You've got a 20% discount.")
      print(f'Please pay {self.bill} dollar(s).')
    else:
      print(f'Please pay {self.bill} dollar(s).')


customer_1 = UniversalStudiosUser("Alice", 21, "Special")
print("--------- 1 ---------")
customer_1.selected_rides("Waterworld-100", "Accelerator-200", "DinoSoarin-50")
print("--------- 2 ---------")
customer_1.show_details()


print("=================")


customer_2 = UniversalStudiosUser("Bob", 20, "Normal")
print("--------- 3 ---------")
customer_2.selected_rides("Enchanted Airways-300", "Jurassic Park-500", "Accelerator-200", "DinoSoarin-50")
print("--------- 4 ---------")
customer_2.show_details()
print("=================")


customer_3 = UniversalStudiosUser("Mark", 15, "Special")
print("--------- 5 ---------")
customer_3.selected_rides("Transformers-450", "Jurassic Park-500", "Waterworld-100", "DinoSoarin-50")
print("--------- 6 ---------")
customer_3.show_details()



```

```python
#task-(University Student managemnet system with OOP):
class Department:
  def __init__(self,dept='ChE Department',sec=5):
    self.dept=dept
    self.sec=sec
    self.avg=0
    self.total=0

    print(f'The {self.dept} has {self.sec} sections.')

  def add_students(self,*args):
    if len(args) == self.sec:
      self.avg = sum(args) / len(args)
      print(f'The {self.dept} has an average of {self.avg} students in each section.')
    else:
      print(f'The {self.dept} doesn"t have {len(args)} sections.')

  def merge_Department(self,*args):
    self.total = self.avg * self.sec
    for dpt in args:
      print(f'{dpt.dept} is merged to {self.dept}.')
      self.total += dpt.avg * dpt.sec
    self.avg = self.total/self.sec
    out=f'Now the {self.dept} has an average of {self.avg} students in each section.'
    return out

d1 = Department()
print('1-----------------------------------')
d2 = Department('MME Department')
print('2-----------------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------------')
print(mega.merge_Department(d3))
```

```python
#task-(Student routine generator with OOP):
class StudentRoutineGenerator:
  def __init__(self,name,max):
    self.name=name
    self.max=max
    self.schedule={'Sat/Thurs': {}, 'Sun/Tue': {}, 'Mon/Wed': {}}
    self.course_count=0
    print(f'Name:{self.name}')
    print(f'Maximum number of courses:{self.max}')
    print(f"Initial Routine: {self.schedule}")

  def addCourses(self,*args):
   if self.course_count <= self.max :
    for item in args:
      new= item.split('-')
      if new[2] not in self.schedule[new[1]]:
        if new[0] not in self.schedule[new[1]].values():
          self.schedule[new[1]][new[2]] = new[0]
          print(f'Successfully added {new[0]}!')
          self.course_count +=1
        else:
          print(f'You already have {new[0]} in your routine')
      else:
        print(f'Can"t take {new[0]}. It"s clashing with your {self.schedule[new[1]][new[2]]}')
   else:
    print(f'You can"t take more than {self.max} courses ')
  def dropCourse(self,*args):
    for old_course in  args:
      for day,inner in self.schedule.items():
        if old_course in inner.values():
          copy_inner=inner.copy()
          for time in copy_inner.keys():
            if old_course == self.schedule[day][time]:
              print(f'Successfully dropped {self.schedule[day][time]}')
              del self.schedule[day][time]
            else:
              print(f'No such course in your routine')
  def showRoutine(self):
    print(f'Updated Routine:')
    print(self.schedule)
    print('Routine Details:')
    for day,inner in self.schedule.items():
      for time,course in inner.items():
        if self.schedule [day][time] != '':
          print(day)
          print(f'{time} - {course}')


print('##################################')
st1 = StudentRoutineGenerator('Harry', 4)
print('------------------------------')
st1.addCourses('CSE110-Mon/Wed-12:30', 'MAT110-Mon/Wed-2:00')
st1.addCourses('ENG101-Sun/Tue-12:30', 'CSE110-Mon/Wed-9:30')
st1.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st1.showRoutine()
print('------------------------------')
st1.dropCourse('CSE110')
st1.dropCourse('PHY112')
print('------------------------------')
st1.showRoutine()
print('##################################')
st2 = StudentRoutineGenerator('John', 3)
print('------------------------------')
st2.addCourses('MAT110-Mon/Wed-8:00')
st2.addCourses('ENG101-Sat/Thurs-12:30', 'CSE110-Sun/Tue-9:30')
st2.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st2.showRoutine()





```
